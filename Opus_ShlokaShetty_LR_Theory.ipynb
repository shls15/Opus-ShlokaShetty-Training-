5.	How is the best fit line chosen?

```python
We choose best fit line by minimizing mean squared error between actual and predicted values using gradient descent and equation.
```

6.	What is gradient descent, and why is it used? Explain the maths.

```python
It is an optimization algorithm that is used to reduce error by iteratively updating paramters to minimize cost function.
It uses derivatives to find slope of cost function and updates values in negative direction of slope.
```

7.	What are residuals?

```python
Residuals are the difference between actual and predicted values.So they help determine how far predictions of model are from true values.
```

8.	What is correlation?

```python
Strength and direction of the relationship between two variables.It ranges between -1 and 1.
```

9.	What is multicollinearity?

```python
It occurs when independent variables are highly correlated.
```

10.	How to detect multicollinearity?

```python
To detect multicollinearity we can use VIF or correlation matrix.VIF values greater than 10 indicate multicollinearity.
```

11.	What are the remedies for multicollinearity?

```python
Remove correlated variables or combine them.Use PCA or Lasso/Ridge regression.
```

12.	What is the R-Squared Statistics?

```python
It shows how well the model explains variance in the datset.Its values lies between 0 and 1.
```

13.	What is an adjusted R-Squared Statistics?

```python
Adjusted R² adjusts R² by considering number of features.It penalizes unnecessary variables.
```

14.	Why do we use adj R-squared?

```python
It gives a more accurate measure when multiple features are used.It prevents overestimation of model performance.
```

15.	Why adj R-squared decreases when we use incompetent variables?



```python
Because it penalizes features that do not improve the model.Irrelevant variables increase complexity without reducing error.
```

16.	How to interpret a Linear Regression model?
