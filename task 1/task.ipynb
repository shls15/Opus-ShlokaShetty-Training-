```python
import scipy.sparse as sp
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python

def build_ui_matrix(df: pd.DataFrame):
    agg = (df.groupby(['CustomerID','StockCode']).size().gt(0).astype(int).reset_index(name='value'))
    users = pd.Categorical(agg['CustomerID'])
    items = pd.Categorical(agg['StockCode'])
    ui = sp.csr_matrix(
        (agg['value'].astype(float), (users.codes, items.codes)),
        shape=(len(users.categories), len(items.categories))
    )
    user_index = pd.Series(range(len(users.categories)), index=users.categories)  
    item_index = pd.Series(range(len(items.categories)), index=items.categories)  
    return ui, user_index, item_index
```

```python

def build_tfidf(df: pd.DataFrame, item_index: pd.Series,
                min_df=2, max_df=0.8, ngram_range=(1,2), stop_words='english'):
  
    items_meta = (df[['StockCode', 'Description']]
                    .drop_duplicates('StockCode')
                    .set_index('StockCode'))

    items_meta = items_meta.reindex(item_index.index)
    items_meta['Description'] = items_meta['Description'].fillna('')

    vectorizer = TfidfVectorizer(min_df=min_df,max_df=max_df,ngram_range=ngram_range,stop_words=stop_words)
    item_content = vectorizer.fit_transform(items_meta['Description'])
    item_content = normalize(item_content)
    return item_content, vectorizer, items_meta
```

```python
def user_profiles_from_transactions(ui: sp.csr_matrix, item_content: sp.csr_matrix):
    ui_bin = ui.copy()
    ui_bin.data = np.ones_like(ui_bin.data)   
    user_profiles = normalize(ui_bin) @ item_content  
    user_profiles = normalize(user_profiles)          
    return user_profiles
```

```python

def recommend_for_user_content_only(user_id: str,ui: sp.csr_matrix,
                                    user_index: pd.Series,
                                    item_index: pd.Series,
                                    items_meta: pd.DataFrame,
                                    user_profiles: sp.csr_matrix,
                                    item_content: sp.csr_matrix,
                                    N: int = 10,
                                    exclude_seen: bool = True,
                                    popularity_fallback: pd.Series | None = None):


    if user_id not in user_index.index:
        return _fallback_top(items_meta, item_index, popularity_fallback, N)
    
    u = user_index.loc[user_id]
    uvec = user_profiles[u]  
    if uvec.nnz == 0:
        return _fallback_top(items_meta, item_index, popularity_fallback, N)
    scores = (uvec @ item_content.T).A1  
    if exclude_seen:
        seen = set(ui[u].indices)
        if seen:
            scores[list(seen)] = -np.inf
    
    op_idx = np.argpartition(-scores, range(min(N, len(scores))))[:N]
    top_idx = top_idx[np.argsort(-scores[top_idx])]
    codes = item_index.index[top_idx]

    out = (items_meta.loc[codes]
           .assign(StockCode=codes, score=scores[top_idx])
           .reset_index(drop=True))
    return out

def _fallback_top(items_meta, item_index, popularity_fallback, N):
    if popularity_fallback is not None:
        ranks = popularity_fallback.to_numpy()
        idx = np.argsort(-ranks)[:N]
    else:
        idx = np.arange(min(N, len(item_index)))
    codes = item_index.index[idx]
    return (items_meta.loc[codes].assign(StockCode=codes, reason='cold-start').reset_index(drop=True))

```

```python

def compute_popularity(df: pd.DataFrame, item_index: pd.Series, by='count'):
    if by == 'count':
        pop = (df.groupby('StockCode').size()
                 .reindex(item_index.index).fillna(0).astype(int))
    elif by == 'revenue':
        pop = (df.assign(line_total=df['Quantity']*df['UnitPrice'])
                 .groupby('StockCode')['line_total']
                 .sum().reindex(item_index.index).fillna(0.0))
    else:
        raise ValueError("by must be 'count' or 'revenue'")
    return pop
```
